```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction import xrd
import matplotlib.pyplot as plt
```

```python
api_key="An68GpwSiKdQlBoZ"
m=MPRester(api_key)
struc = m.get_structure_by_material_id("mp-2133", False)
```

```python
xrayd = xrd.XRDCalculator(wavelength=1.5406, symprec=0, debye_waller_factors=None)
pattern = xrayd.get_pattern(struc, scaled=True, two_theta_range=(0,90))
pattern
```

```python
theta= pattern.x
intensity = pattern.y
hkil = pattern.hkls
```

```python
xrayd.show_plot(struc)
print("Dans le cas des structures  symtrie hexagonale, ou trigonale,", 
      "on dfinit parfois un quatrime indice pour dsigner les plans, (hkil) ;",
      "c'est la notation de Bravais-Miller. L'indice i, plac en troisime position,",
      "est redondant (les trois indices h, k et l suffisent  eux seuls  dfinir un plan) ;",
      "il est dfini par i = -h - k.")
print("Cette notation permet d'appliquer des permutations circulaires d'indices",
      "pour dfinir des familles de plans.")
print("Source: https://fr.wikipedia.org/wiki/Indices_de_Miller_et_indices_de_direction#Cristal__symtrie_hexagonale")
print('')
print("L'echelle d'intensit est dite scaled, c'est--dire que l'intensit la plus haute, I(max), ",
      "aura la valeur 100, les autres intensits respecteront la relation I(scaled) = 100*I/I(max)")
print('')
print('')

print("PREMIER PIC:")
print("le premier pic apparait pour un angle 2 theta =", theta[0], '')
print("Son intensit est de", intensity[0])
print("Ses indices hkil et sa multiplicit assocs sont", hkil[0])
print("Ses indices hkl associs sont donc [1,0,0]")
print('')

print("DEUXIEME PIC:")
print("le deuxieme pic apparait pour un angle 2 theta =", theta[1], '')
print("Son intensit est de", intensity[1])
print("Ses indices hkil et sa multiplicit assocs sont", hkil[1])
print("Ses indices hkl associs sont donc [0,0,2]")
print('')

print("TROISIEME PIC:")
print("le troisieme pic apparait pour un angle 2 theta =", theta[2], '')
print("Son intensit est de", intensity[2])
print("Ses indices hkil et sa multiplicit assocs sont", hkil[2])
print("Ses indices hkl associs sont donc [1,0,1]")
print('')
```
